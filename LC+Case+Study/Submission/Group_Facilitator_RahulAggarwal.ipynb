```python
#binning annual income
# annual income
def annual_income(n):
    if n <= 50000:
        return 'low'
    elif n > 50000 and n <=100000:
        return 'medium'
    elif n > 100000 and n <=150000:
        return 'high'
    else:
        return 'very high'
dfloan['annual_inc'] = dfloan['annual_inc'].apply(lambda x: annual_income(x))
```

```python
def funded_amount(n):
    if n <= 5000:
        return 'low'
    elif n > 5000 and n <=15000:
        return 'medium'
    else:
        return 'high'
    
dfloan['funded_amnt'] = dfloan['funded_amnt'].apply(lambda x: funded_amount(x))
```

```python
varPlot('annual_inc')
```

**Low income group tends to be more defaulter**

```python
def emp_length(n):
    if n <= 1:
        return 'fresher'
    elif n > 1 and n <=3:
        return 'mid Level'
    elif n > 3 and n <=7:
        return 'Higher Level'
    else:
        return 'Top Level'

dfloan['emp_length'] = dfloan['emp_length'].apply(lambda x: emp_length(x))
```

```python
varPlot('emp_length')
```

**As per the graph employment length is not a good discriminator**

```python
def dti(n):
    if n <= 10:
        return 'low'
    elif n > 10 and n <=20:
        return 'medium'
    else:
        return 'high'
    

dfloan['debt_to_incm'] = dfloan['debt_to_incm'].apply(lambda x: dti(x))
```

Observation:

    
1. loan_amnt is correlated to last_payment_amount with r factor.48, as expected
2. int_rate is highly correlated to revol_util with r factor of .47, which shows that company is charging the defaulter with higher rate of interest.
3. int_rate is highly correlated with grade_rank, as we have also seen in graphs above.
4. loan_amnt revol_bal are correlated with r factor .32 - This is not good as it suggests that higher loan amount is being approved to riskier borrowers.

6. revol_util is inversly or un-correlated with earliest_cr_line_year. We depict that these two features are independent.
