```python
master_frame=master_frame.drop('funding_round_code', axis=1)
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

*Checking the rows which have more than 5 values as null*

```python
master_frame[master_frame.isnull().sum(axis=1) > 5]
```

```python
#sum of the values for more than 5 which are null 
len(master_frame[master_frame.isnull().sum(axis=1) > 5].index)
```

```python
#percentage of values for more than 5 which are null  
100*(len(master_frame[master_frame.isnull().sum(axis=1) > 5].index) / len(master_frame.index))
```

```python
# removing NaN raised_amount_usd rows
master_frame = master_frame[~np.isnan(master_frame['raised_amount_usd'])]
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
#retaining the rows for which less than 5 values rae null or NaN
master_frame = master_frame[master_frame.isnull().sum(axis=1) <= 5]
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

deleting **founded_at** as it is not required in current analysis

```python
master_frame=master_frame.drop('founded_at', axis=1)
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
master_frame['country_code'] = master_frame['country_code'].astype('category')
master_frame['country_code'].value_counts()
```

```python
#The most common value of ciuntry_code is USA so imputing the null values with USA
master_frame.loc[pd.isnull(master_frame['country_code']), ['country_code']] = 'USA'
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
master_frame['state_code'] = master_frame['state_code'].astype('category')
master_frame['state_code'].value_counts()
```

```python
#The most common value of state_code is CA so imputing the null values
master_frame.loc[pd.isnull(master_frame['state_code']), ['state_code']] = 'CA'
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
master_frame['region'] = master_frame['region'].astype('category')
master_frame['region'].value_counts()
```

```python
#Most common region is SF Bay area so imputing the null values
master_frame.loc[pd.isnull(master_frame['region']), ['region']] = 'SF Bay Area'
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
#homepage_url category is mot useful in current analysis so dropping the values 
master_frame = master_frame.drop('homepage_url', axis=1)
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

**_Plotting the Graphs_**<br>
A plot showing the fraction of total investments (globally) in angel, venture, seed, and private equity, and the average amount of investment in each funding type. This chart should make it clear that a certain funding type (FT) is best suited for Spark Funds.

A plot showing the top 9 countries against the total amount of investments of funding type FT. This should make the top 3 countries (Country 1, Country 2, and Country 3) very clear.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_investType= master_frame[master_frame['funding_round_type'].isin(['angel','venture', 'seed','private_equity'])]
df_investType
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(x='funding_round_type', y='raised_amount_usd', data=df_investType , estimator=np.mean)
plt.xlabel('Funding Type')
plt.ylabel('Raised Amount in USD (in Millions)')
plt.title('Funding Type Investements')
plt.show()
```

```python
top9=mf_venture.groupby('country_code')['raised_amount_usd'].sum().sort_values(ascending=False)
top9.head(9)

```

```python
top9=top9.to_frame()
top9=top9.head(9)
top9
```

```python
top9.reset_index(level=0,inplace=True)
top9
```

```python
top9['country_code'] = top9['country_code'].astype('object')
```

```python
plt.figure(dpi=100 , figsize=(10, 5))
sns.barplot(x='country_code', y='raised_amount_usd', data=top9 , estimator=np.sum ,palette="Blues_d")
plt.xlabel('Countries')
plt.ylabel('Total Amount of Investments')
plt.yscale('log')
plt.title('Funding Analysis on Countries')
plt.show()
```
